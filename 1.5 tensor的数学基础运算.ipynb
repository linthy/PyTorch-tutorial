```python
import torch
```

```python
a = torch.rand(4,4)
b = torch.rand(4)
```

```python
a+b,torch.add(a,b)
```

```python
a-b,torch.sub(a,b)
```

```python
a*b,torch.mul(a,b)
```

```python
a/b,torch.div(a,b)
```

```python
c = torch.rand(4,3,28,32)
d = torch.rand(4,3,32,64)
```

```python
# 注意是 @ 符号 
torch.all(torch.eq(torch.matmul(c,d),c@d))
```

```python
c@d , torch.matmul(c,d)
```

#### 数值范围处理

```python
e = torch.rand(1) * 10
e
```

```python
# 下取值，上取值，四舍五入取值
torch.floor(e),torch.ceil(e),torch.round(e)
```

```python
# 数值分解成 整数和小数  truncate and  fraction
torch.trunc(e), torch.frac(e)
```

```python
# 设定上下限 clamp
f = torch.rand(3,3)*15
print(f)
f.clamp(5,10)
```

```python
f.where(f>5,torch.tensor(5.))
```
